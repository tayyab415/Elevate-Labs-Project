```python
url = 'https://in.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords=&location=india&start=1'

response = requests.get(url)

list_data = response.text
list_soup = BeautifulSoup(list_data, 'html.parser')
page_jobs = list_soup.find_all('li')
```

```python
id_list = []
for job in page_jobs:
    base_card_div = job.find("div", {"class": "base-card" })
    job_id = base_card_div.get("data-entity-urn").split(":")[3]
    print(job_id)
    id_list.append(job_id)
```

```python

# It's a good practice to use headers to mimic a browser request and avoid being blocked.
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

for job_id in id_list:
    # Adding a random delay between requests to be respectful to the server and reduce the chances of being blocked.
    time.sleep(random.uniform(1, 4))
    
    job_url = f'https://in.linkedin.com/jobs-guest/jobs/api/jobPosting/{job_id}'
    
    try:
        job_response = requests.get(job_url, headers=headers, timeout=10)
        
        print(f"Requesting ID {job_id}... Status: {job_response.status_code}")
        
        # A simple retry mechanism: if we are rate-limited (429), wait and try one more time.
        if job_response.status_code == 429:
            print("We are being rate-limited. Waiting 30 seconds...")
            time.sleep(30)
            job_response = requests.get(job_url, headers=headers, timeout=10)
            print(f"Retrying ID {job_id}... Status: {job_response.status_code}")
            
        job_post = {}
        if job_response.status_code == 200:
            job_soup = BeautifulSoup(job_response.text, 'html.parser')
            
            # Using select_one with CSS selectors is often more robust than finding by a long, exact class string.
            job_post['job_title'] = job_soup.select_one('.top-card-layout__title').text.strip() if job_soup.select_one('.top-card-layout__title') else None
            job_post['company_name'] = job_soup.select_one('.topcard__org-name-link').text.strip() if job_soup.select_one('.topcard__org-name-link') else None
            job_post['time_posted'] = job_soup.select_one('.posted-time-ago__text').text.strip() if job_soup.select_one('.posted-time-ago__text') else None
            job_post['location'] = job_soup.select_one('.topcard__flavor--bullet').text.strip() if job_soup.select_one('.topcard__flavor--bullet') else None
            job_post['number_of_applicants'] = job_soup.select_one('.num-applicants__caption').text.strip() if job_soup.select_one('.num-applicants__caption') else None
            job_post['job_function'] = job_soup.select_one('.description__job-criteria-text--criteria').text.strip() if job_soup.select_one('.num-applicants__caption') else None
        
        job_list.append(job_post)

    except requests.exceptions.RequestException as e:
        print(f"An error occurred for job ID {job_id}: {e}")
        # Append empty data if the request fails
        job_list.append({})
```

```python
job_df = pd.DataFrame(job_list)
job_df.tail(20)
```

```python
url = 'https://in.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords=&location=india&start=41'

response = requests.get(url)

list_data = response.text
list_soup = BeautifulSoup(list_data, 'html.parser')
page_jobs = list_soup.find_all('li')
```

```python
id_list = []
for job in page_jobs:
    base_card_div = job.find("div", {"class": "base-card" })
    job_id = base_card_div.get("data-entity-urn").split(":")[3]
    print(job_id)
    id_list.append(job_id)
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
# Adding the time and random imports if they are not in the first cell
import time
import random

for job_id in id_list:
    # Adding a random delay to be respectful and avoid getting blocked
    time.sleep(random.uniform(1, 4))
    job_url = f'https://in.linkedin.com/jobs-guest/jobs/api/jobPosting/{job_id}'
    
    try:
        job_response = requests.get(job_url, headers=headers, timeout=10)
        print(f"Requesting ID {job_id}... Status: {job_response.status_code}")

        if job_response.status_code == 429:
            print("Rate limit hit. Waiting for 30 seconds.")
            time.sleep(30)
            job_response = requests.get(job_url, headers=headers, timeout=10)
            print(f"Retrying ID {job_id}... Status: {job_response.status_code}")

        job_post = {'job_id': job_id}
        
        if job_response.status_code == 200:
            job_soup = BeautifulSoup(job_response.text, 'html.parser')
            
            # --- Extracting Basic Info ---
            job_post['job_title'] = job_soup.select_one('.top-card-layout__title').get_text(strip=True) if job_soup.select_one('.top-card-layout__title') else None
            job_post['company_name'] = job_soup.select_one('.topcard__org-name-link').get_text(strip=True) if job_soup.select_one('.topcard__org-name-link') else None
            job_post['time_posted'] = job_soup.select_one('.posted-time-ago__text').get_text(strip=True) if job_soup.select_one('.posted-time-ago__text') else None
            job_post['location'] = job_soup.select_one('.topcard__flavor--bullet').get_text(strip=True) if job_soup.select_one('.topcard__flavor--bullet') else None
            
            # --- Dynamically Extracting All Job Criteria ---
            criteria_items = job_soup.select('li.description__job-criteria-item')
            for item in criteria_items:
                header_element = item.select_one('h3.description__job-criteria-subheader')
                value_element = item.select_one('span.description__job-criteria-text--criteria')
                
                if header_element and value_element:
                    # Clean the header text to use as a dictionary key
                    key = header_element.get_text(strip=True).lower().replace(' ', '_')
                    value = value_element.get_text(strip=True)
                    job_post[key] = value

        job_list.append(job_post)

    except requests.exceptions.RequestException as e:
        print(f"An error occurred for job ID {job_id}: {e}")
        job_list.append({'job_id': job_id, 'error': str(e)})
```

```python
job_df = pd.DataFrame(job_list)
job_df.tail(20)
```
